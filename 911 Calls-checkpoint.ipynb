<h4>In the 'title' column there are 'Reasons/Departments' specified before the title code. These are EMS, Fire, and Traffic. I will create a new column called 'Reason' that contains this string value</h4>

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason']
```

<h4> What is the most common reason for 911 calls? </h4>

```python
df['Reason'].value_counts().head(1)
```

<h4>Use seaborn to create a countplot of 911 calls  by Reason</h4>

```python
sns.countplot(x='Reason', data = df)
plt.show()
```

<h4>Create 4 new Columns called 'Hour','Month','Year' and 'Day of the week' based off of the timeStamp column</h4>

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Year'] = df['timeStamp'].apply(lambda time: time.year)
df['Day of the week'] = df['timeStamp'].apply(lambda time: time.day_of_week)
```

```python
df.head()
```

<h4>Convert days of the week from numbers to the actual days</h4>

```python
dmap = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
df['Day of the week'] = df['Day of the week'].map(dmap)
```

```python
df.head()
```

<h4>Create countplot of the 'Day of the week' column with the hue based off of the 'Reason' column</h4>

```python
sns.countplot(x='Day of the week',data = df)
plt.show()
```

```python
sns.countplot(x='Day of the week',data = df, hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

<h4>Create countplot of the 'Month' column with the hue based off of the 'Reason' column</h4>

```python
sns.countplot(x='Month',data = df)
plt.show()
```

```python
sns.countplot(x='Month',data = df, hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

<h4>Create a new column called 'Date' that contains the data from the 'timeStamp' column</h4>

```python
df['Date'] = df['timeStamp'].apply(lambda x:x.date())
```

```python
df.head()
```

<h4>Group by the 'Date' column and create a plot of counts of 911 calls within a year</h4>

```python
df[df['Year']==2016].groupby('Date').count()['lat'].plot()
plt.tight_layout()
```

```python
df[(df['Reason'] == 'Traffic') & (df['Year'] == 2016)].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[(df['Reason'] == 'Fire') & (df['Year'] == 2016)].groupby('Date').count()['lat'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[(df['Reason'] == 'EMS') & (df['Year'] == 2016)].groupby('Date').count()['lat'].plot()
plt.title('EMS')
plt.tight_layout()
```

<h4>Create a heatmap and clustermap for the number of calls per hour per day of the week</h4>

```python
dayHour = df.groupby(['Day of the week','Hour']).count()['Reason'].unstack()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
dayHour
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour, cmap='viridis')
plt.show()
```

```python
sns.clustermap(dayHour, cmap='coolwarm')
plt.show()
```

<h4>Create a heatmap and clustermap for the number of calls per Month per day of the week</h4>

```python
dayMonth = df.groupby(['Day of the week','Month']).count()['Reason'].unstack()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
dayMonth
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth, cmap='viridis')
plt.show()
```

```python
sns.clustermap(dayMonth, cmap='coolwarm')
plt.show()
```
